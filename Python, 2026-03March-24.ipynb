# Nouns vs. verbs

In the world of programming, we often divide things into nouns (data) and verbs (functions):

- Nouns are strings, integers, dictionaries, etc.
- Verbs would be functions and methods

We've seen over the last few weeks how we can define new nouns -- data structures that contain values we want, and even combinations of data structures. Each data structure we create allows us to think about and organize our values in a new way.

Verbs are really where it's at. We can define a new verb, a new function.

Does that really add new functionality to our program? NO. We can't do anything new with functions defined that we couldn't do before. But we can think at a higher level, and not have to worry about low-level details.

This is an example of abstraction, an important idea in engineering and especially in software. The idea is that you collect a bunch of things together, give it a name, and treat it as one thing. That allows you to build on top of that collection more easily, because there's less to keep in mind.\

When we define a new function, we're collecting a number of previously defined actions and data structures into a single name and verb. We can then invoke that function, and think at a higher level.

We can also write another new function that uses our previously defined function. In this way, we create a pyramid of functions, where we're always thinking at higher and higher levels.

We can define functions in Python, and that's one of the most important and common things that programmers do. HOWEVER, in Python, functions aren't just verbs. They're also nouns. When we define a function, we're actually creating a new data structure.

- After the name, we put `()`. This where the *parameters* go, the variables that will accept the arguments when the function is called. For starters, we're going to leave these blank.

The function body does *not* execute when we define the function! It only runs when we *call* or *invoke* the function.

1. Define a function, `calc`, that when invoked asks the user three questions:
    - What is the first number?
    - What is the operator?
    - What is the second number?
2. Assign the answer from each question to a variable.
3. If the numbers can be turned into integers (checking with `str.isdigit`), and if the operator is known (`+` or `-`), then get the result, and print the full expression for the user.
4. If the numbers cannot be turned into integers, then scold the user.
5. If the operator isn't known, then make the result `(no result)`.

Example:

    calc()
    
    First: 7
    Operator: +
    Second: 8
    7 + 8 = 15

```python
def calc():
    first = input('First: ').strip()
    operator = input('Operator: ').strip()
    second = input('Second: ').strip()

    if first.isdigit() and second.isdigit():
    
        first = int(first)
        second = int(second)
    
        if operator == '+':
            result = first + second
        elif operator == '-':
            result = first - second
        else:
            result = '(No result)'
    
        print(f'{first} {operator} {second} = {result}')

    else:
        print(f'Make sure that {first} and {second} are both numeric')
```

# Return values

So far, our functions have used `print` for their output. To the casual observer, that's fine. But in reality, it's bad -- we don't want to have the function's output printed on the screen. Rather, we want it to be *returned* from the function. If it's returned, then we can capture the value in variable and use it moving forward.

Every function returns a value. If you don't explicitly return one, then the function returns `None`. Every function can return *one* value, but it can be of any Python type.

To return a value, use the `return` keyword in your function. As soon as the function hits `return`, the value is returned to the caller and the function's execution stops.

If the function's execution gets to the end of the function body, then it automatically returns the `None` special value.



```python
def hello(name):
    return f'Hello, {name}!'
```

```python
hello('world')  # in Jupyter, we see return values -- here, the string with ''
```

```python
x = hello('world')  # nothing is displayed, because we captured the value in x
```

```python
print(x)
```

```python
for i in range(5):
    print(x)
```

# How does anyone know?

- How does someone know how many arguments they should pass to our function?
- How do they know what types of values we expect?
- How do they know what a function even does?
- How do they know what the function returns?

All of this is put in a *docstring*. That's a special string, one that goes in the first line of the function, and isn't assigned to a variable. 

Traditionally, we use a triple-quoted string for the docstring, allowing us to have more than one line.

The docstring is *not* a comment! Rather, the docstring is for who will be *using* our function, and wants to know how to invoke it.

```python
def hello(first, last):
    """Returns a string with a friendly greeting.

    Expects: Two arguments, both strings
    Modifies: Nothing
    Returns: A string with a friendly greeting incorporating both names
    """

    return f'Hello, {first} {last}!'
```

```python
hello('Reuven', 'Lerner')
```

# How can people see the docstring?

- In Jupyter, we can invoke the `help` function, passing the `hello` function name as an argument.
- In editors like PyCharm and VSCode, you can hover over the function name, and it'll show you the docstring.

```python
help(hello)  # I'm passing hello as an argument
```

# Keyword arguments

Sometimes, we don't want to use positional arguments. Maybe it's hard to remember which argument goes to which parameter. Maybe we want to mess with the order. 

For whatever reason, you can invoke a function, passing *keyword arguments*. Positional arguments are assigned to parameters based on their positions. Keyword arguments explicitly indicate which parameter should get the value.

keyword arguments are always written as `name=value`, with the parameter name on the left and the value on the right.

```python
# we can use keyword arguments with nearly any function

def add(x, y):
    total = x + y
    return total

    
```

```python
add(10, 3)   #positional arguments
```

```python
# parameters:   x   y
# arguments:   10   3   # these are assigned by name

add(x=10, y=3)  # now we're using keyword arguments
```

```python
# can I call them with a different order? Answer: YES!

# parameters:   x   y
# arguments:    10    3

add(y=3, x=10)
```

```python
# can I mix positional and keyword arguments?
# yes, but all positional must come before all keyword

add(10, y=3)   # this is fine!
```

```python
add(x=10, 3)   # this is not fine!
```

# When you're calling a function:

- All positional arguments come first
- All keyword arguments come next

    - `text`, a string
    - `to_count`, another string
2. The function will count how many times one of the characters from `to_count` is in `text`.
3. If we invoke the function with `to_count='aeiou'`, then it'll tell us how many vowels are in `text`. If we invoke it with `to_count='abcde'`, then it'll count those characters.
4. The return value will be an integer, the count

Invoke our function in three ways:
1. With 2 positional arguments
2. With 2 keyword arguments
3. With 1 of each

Example:

    count_characters('hello out there', 'aeiou')   # return 6

```python
def count_characters(text, to_count):
    total = 0

    for one_character in text:         # go through the text, one character at a time
        if one_character in to_count:  # if this is a character we want to count...
            total += 1                 # add it to total

    return total
```

# Default argument values

We've seen a number of examples of methods and functions that can be invoked with varying numbers of arguments. How is that possible? So far, all of our functions need to be invoked with a number of arguments matching the number of parameters.

This is possible with *default argument values*. In other words: If we pass an argument, then the argument is assigned to the parameter. But if we don't pass an argument, then a default value is assigned to that parameter. That default is defined along with the function definition.

```python
def add(x, y):
    total = x + y
    return total
```

```python
add(10, 3)
```

```python
add(10)  # error - -missing one argument
```

```python
# now we'll give y a default argument value

def add(x, y=5):
    total = x + y
    return total
```

```python
# parameters: x   y
# arguments:  10   3

add(10, 3)
```

```python
# parameters: x     y
# arguments:  10   5

add(10)
```

# Rules for default argument values

1. All mandatory parameters (i.e., without default argument values) must come before optional parameters (i.e., with default argument values).
2. You should never use mutable data (e.g., lists and dicts) as a default argument value.
