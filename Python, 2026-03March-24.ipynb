### to open a file for reading, I give a filename as a string

- If I just say `open('filename.txt')`, then Python looks in the same directory as the program. If you're using Jupyter, then in the same folder from which you ran Jupyter.
- If you're running Jupyter lite in your browser, then it only looks in the pretend directories on the left panel of your browser. It does not and cannot access any of the files on your actual computer. This does mean that in order to see the file and open it, you need to drag it from your computer's desktop to that left-side panel in your browser.

### If the filename you give includes a `/` (on Unix/MacOS) or `\\` on Windows, then:

- That's known as a "relative path," and it goes deeper down from where the program was run (into a subfolder/subdirectory) or higher up on the disk.

### If the filename *starts* with `/` (on Unix/MacOS) or `C:\` on Windows, then:

That's known as an "absolute path," and doesn't depend on where you're running the program from. When I open `/etc/passwd`, it doesn't matter where my program is, because that's an absolute path, startign with `/`, the "root" directory in Unix.

# Nouns vs. verbs

In the world of programming, we often divide things into nouns (data) and verbs (functions):

- Nouns are strings, integers, dictionaries, etc.
- Verbs would be functions and methods

We've seen over the last few weeks how we can define new nouns -- data structures that contain values we want, and even combinations of data structures. Each data structure we create allows us to think about and organize our values in a new way.

Verbs are really where it's at. We can define a new verb, a new function.

Does that really add new functionality to our program? NO. We can't do anything new with functions defined that we couldn't do before. But we can think at a higher level, and not have to worry about low-level details.

This is an example of abstraction, an important idea in engineering and especially in software. The idea is that you collect a bunch of things together, give it a name, and treat it as one thing. That allows you to build on top of that collection more easily, because there's less to keep in mind.\

When we define a new function, we're collecting a number of previously defined actions and data structures into a single name and verb. We can then invoke that function, and think at a higher level.

We can also write another new function that uses our previously defined function. In this way, we create a pyramid of functions, where we're always thinking at higher and higher levels.

We can define functions in Python, and that's one of the most important and common things that programmers do. HOWEVER, in Python, functions aren't just verbs. They're also nouns. When we define a function, we're actually creating a new data structure.
