# What are modules?

Remember the DRY rule? Don't Repeat Yourself!

- If we have the same code on several consecutive lines, we can consolidate them into a single `for` loop.
- If we have the same code in several places within our program, we can define a function and then invoke the function multiple times.
- If we have the same code in several places in *different* programs, then we can make use of a *library*.

(Almost) all programming languages off a library, meaning functionality that we can load when we need or want it.

In Python, our libraries are called "modules" or "packages."  A module is a single file of Python code that we can load, and a package is a directory/folder of modules that we can load, either all together or one at a time.

Modules in Python do more than just store data and functions that we might want to use in multiple programs. They are also *namespaces*. The basic thinking is that if you want to use the variable `x` in your part of a program, and I want to use `x` in my part of a program, and we want to have our parts work together, then there might be problems.

Thanks to modules, this largely becomes a non-issue, so long as the two parts of our program are defined in separate modules.

# Using `import`

- It's a statement, not a function, Just write `import`, no `()` after the name
- The name we give after `import`, the variable, is a bit weird. Yes, it's a variable name. But it also describes the filename that Python should look for. In this case, it's going to look for `random.py`.
- Unlike many other programming languages, Python doesn't allow us to give a filename to `import`. Rather, Python needs to be told where to look.
- Immediately after `import`, you have access to the module's functionality.
- Typically, the module will contain one or more functions, and one or more data structures. All are accessed via attributes -- meaning, names that come after the `.`

# Exercise: Guessing game

1. Have the computer choose a random integer from 0-100. Use `random.randint` and assign to `secret`.
2. Ask the user, repeatedly, to guess a number:
    - If it's too high, say "too high"
    - if it's too low, say "too low"
    - if it's right, say "just right" and exit
3. If it's not right, then keep going, and ask the user to guess again.
4. When exiting the program, print not only the number the user correctly guessed, but also how many guesses it took.

```python
import random

number = random.randint(0, 100)
guess_counter = 0

while True:
    s = input('Guess: ').strip()
    
    if not s.isdigit():
        print(f'{s} is not numeric; try again')
        continue
        
    guess = int(s)
    guess_counter += 1
    
    if guess == number:
        print('You got it!')    
        break
    elif guess < number:
        print('Too low!')
    else:
        print('Too high!')
    
print(f'You guessed {number} in {guess_counter} tries.')
```

# `__name__`

`__name__` is defined everywhere in Python. It reflects the current *namespace*. Normally, a module that is imported will have `__name__` defined to be the name of the file (without the `.py` extension).

What happens if we execute our module not via `import`, but just run it on the command line?

If we run the program from the command line, `__name__` is defined to be the special string `'__name__'`. 

In a large Python program, you can imagine that there will be hundreds or even thousands of modules imported. Each will have `__name__` defined to be its filename. All, except for one -- the first program that is executed directly, rather than imported.

We can use this to distinguish between when a module has been imported, and when it is being run as a program. This lets us produce different output for different circumstances.

```python
importlib.reload(mymod)
```

# About that `if` statement

The line, `if __name__ == '__main__'` is one of the most common, famous lines in all of Python! All it means is:

- If this module is being run as a standalone program, and was not imported, then do the things below this line.

This is *not* a mandatory line, and it has nothing to do with the "main" function in C and other languages.

# Where do modules come from?

We've seen already that there are two basic sources:

- Modules that come with Python
- Modules that we write

Both have to be in `sys.path`.

"Modules that come with Python" are known as the "standard library."  The standard library is included when you download and install Python. You can be sure that if a module is in the standard library, then it's available on every computer with Python.

This means that if you use such a module, and then you distribute your Python program to other users, they'll have the module you referenced -- they don't need to install anything more.

Because the standard library comes with Python, it's only updated when a new version of Python is released. (Major versions happen in October, and minor versions throughout the year.) If a bug is found in the standard library, it'll only get fixed when the next Python version comes out -- unless it's a very serious bug, in which case a new version of Python will be pushed ASAP.

Let's look at the standard library!

# Exercise: Globbing

The `glob` module contains a function `glob` (so yes, `glob.glob`), which takes a single argument, a string containing a pattern of filenames. If you've ever used `*.txt`, then you've used globbing before. In globbing, `*` means "any number of any characters" and `?` means "one character of any type."

1. `import glob`
2. Ask the user to enter a pattern of filenames
3. Find eachof the files matching this pattern
4. For each file, print its size (which you can get by opening the file and totalling the lengths of the lines)
5. The output should be filenames and their sizes

To avoid problems, do try to give a pattern that points to text files, including Python files. So something like `*.py` would be great, with all of the Python files in the current directory.

```python
import glob

glob.glob('*.py')
```

```python
glob.glob('*.txt')
```

```python
for one_filename in glob.glob('*.txt'):
    file_length = len(open(one_filename).read())
    print(f'{one_filename}: {file_length}')
```

```python
import glob

pattern = input('Enter globbing pattern: ').strip()

for one_filename in glob.glob(pattern):
    file_length = len(open(one_filename).read())
    print(f'{one_filename}: {file_length}')
```

# Installing modules from PyPI

The hard, bad, old way to install a module is to go to the project page on PyPI, and click on "downloadable files," then choose either the source distribution or a wheelfile. Now you can install it on your computer with the `pip` command, for the Python Installer for Packages.

But why work so hard?

You can instead just say

    pip install rich

pip (which comes with Python) will go to PyPI, find the package named "rich", and then download the appropriate file and install it.

Remember that `pip` is *not* a Python function or command! If you want to execute it, you need to be on the command line.
