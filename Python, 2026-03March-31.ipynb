# What are modules?

Remember the DRY rule? Don't Repeat Yourself!

- If we have the same code on several consecutive lines, we can consolidate them into a single `for` loop.
- If we have the same code in several places within our program, we can define a function and then invoke the function multiple times.
- If we have the same code in several places in *different* programs, then we can make use of a *library*.

(Almost) all programming languages off a library, meaning functionality that we can load when we need or want it.

In Python, our libraries are called "modules" or "packages."  A module is a single file of Python code that we can load, and a package is a directory/folder of modules that we can load, either all together or one at a time.

Modules in Python do more than just store data and functions that we might want to use in multiple programs. They are also *namespaces*. The basic thinking is that if you want to use the variable `x` in your part of a program, and I want to use `x` in my part of a program, and we want to have our parts work together, then there might be problems.

Thanks to modules, this largely becomes a non-issue, so long as the two parts of our program are defined in separate modules.





# Using a module

If we want to use a module in Python, we must first load it into memory. We do that with the `import` statement.

Note: `import` is *not* a function, so we don't use `()` after its name. It's actually closest to `def`, in that it's a statement and also the name put after `import` is defined as a variable.

When I import a module, I'm defining a new variable -- you can think of it as a warehouse for other values, where they are stored.

One of my favorite modules to demonstrate with is `random`. It contains a number of functions that help us to choose random values.

```python
import random
```

```python
# having imported random, I now have a "random" variable defined

random
```

```python
random.randint(0, 100)  # this invokes the 'randint' function in random
```

# Using `import`

- It's a statement, not a function, Just write `import`, no `()` after the name
- The name we give after `import`, the variable, is a bit weird. Yes, it's a variable name. But it also describes the filename that Python should look for. In this case, it's going to look for `random.py`.
- Unlike many other programming languages, Python doesn't allow us to give a filename to `import`. Rather, Python needs to be told where to look.
- Immediately after `import`, you have access to the module's functionality.
- Typically, the module will contain one or more functions, and one or more data structures. All are accessed via attributes -- meaning, names that come after the `.`

# Exercise: Guessing game

1. Have the computer choose a random integer from 0-100. Use `random.randint` and assign to `secret`.
2. Ask the user, repeatedly, to guess a number:
    - If it's too high, say "too high"
    - if it's too low, say "too low"
    - if it's right, say "just right" and exit
3. If it's not right, then keep going, and ask the user to guess again.
4. When exiting the program, print not only the number the user correctly guessed, but also how many guesses it took.

```python
import random

number = random.randint(0, 100)
guess_counter = 0

while True:
    s = input('Guess: ').strip()
    
    if not s.isdigit():
        print(f'{s} is not numeric; try again')
        continue
        
    guess = int(s)
    guess_counter += 1
    
    if guess == number:
        print('You got it!')    
        break
    elif guess < number:
        print('Too low!')
    else:
        print('Too high!')
    
print(f'You guessed {number} in {guess_counter} tries.')
```

# Where did `random` come from?

If we ask the `random` module, it'll tell us where it was loaded from.


```python
random
```

How did `import random` know to look there? 

The answer is `sys.path`. This is an attribute (`path`) in the `sys` module that is a list of strings -- each string being a directory name. When you say `import ABCD`, Python runs a `for` loop over `sys.path`, looking for `ABCD.py` in each directory. The first directory in which it can be found wins.



```python
import sys

sys.path
```

Before looking in any of these places in `sys.path`, Python first looks in the current directory. If you're in Jupyter and have a file in the same directory as your notebook, then it'll be loaded as a module.

# Can't I just invoke `randint`?

No. You can't. `randint` is only defined within the `random` module's namespace. You can think of a namespace as a surname for a bunch of variables. 

```python
randint(0, 100)   # Python has no idea what I'm talking about!
```

# I really wanna!

What if you are going to be invoking `random.randint` a lot, and you want to cut down on your typing? Maybe you can get a shorter name than this.

That's what `from .. import` is all about: It lets you indicate which names from a module you want to load into the main namespace, with global variables. No longer will the name be under a module name, but rather on its own.

```python
# this (1) loads the module into memory, unless it was already
# (2) it doesn't define random as a variable
# (3) It does define randint as a global variable

from random import randint
```

```python
randint(0, 100)
```

```python
# you might need or want to say

import random                # loads the whole module, defines the "random" name
from random import randint   # only defines "randint" to be a global variable
```

# What if we want to give the module an alias?

Instead of loading the module and defining a variable named for its filename, maybe we can/should give it an alias. 

This is especially common in the data world, where we say `import pandas as pd` or `import numpy as np`.

The syntax is indeed

    import MODNAME as ALIAS

```python
import random as ra
```

```python
ra.randint(0, 100)
```

```python
from random import randint as ri
```

```python
ri(0, 100)
```

# Four ways to import modules

1. `import MODNAME`
2. `import MODNAME as ALIAS`
3. `from MODNAME import NAME`
4. `from MODNAME import NAME as ALIAS`

```python
from random import choice as ch

months = 'Jan Feb Mar Apr May Jun Jul Aug Sep Oct Nov Dec'.split()
random_month = ch(months)
guess_counter = 0

while True:
    guess = input('Guess the month: ').strip()

    if guess in months:
        guess_counter += 1

    if guess == random_month:
        print('You got it!')
        break      

print(f'You guessed {random_month} in {guess_counter} guesses.')        
```
