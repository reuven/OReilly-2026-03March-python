# Where do we stand?

- Python has lots of values, and each value has a type
- We can assign these values to variables
- Each type has a different reason for being, and advantages vs. disadvantages
    - Numbers (ints and floats) are good for calcuations
    - Strings are good for text
    - Lists are good for sequences of values (a bunch of numbers, a bunch of strings, a bunch of lists)
    - Tuples are good for records of values (several differnet fields, each of a different type)

Today, we'll talk about *dictionaries*, aka *dicts*. A dictionary is not unique to Python, but it is very very useful, because it doesn't store individual values. Rather, it stores pairs of values. One element in that pair is known as the *key* and the other is known as the *value*.

One way to think about dicts is that they're like lists, except that you can dictate the index! You aren't stuck with 0, 1, 2, 3, etc.

Dicts are the most important data structure in Python -- not only because we love to use them, but because the Python language uses them a lot.



# Dicts

A dictionary is Python's term for something that's very famous and well known in the programming world. It usually has other names, though:

- Hash table
- Hash map
- Map
- Name-value pair
- Key-value pair
- Associative array

The idea in a dict is that you aren't storing a single value. Rather, you're storing a *key* (the index, basically) and the value.

# Rules for a dict

- You create a dict with `{}`. This is the empty dict.
- If you want, you can create the dict with one or more key-value pairs.
    - `{'a':10}`
    - `{'a':10, 'b':20}`
- As you can see, we separate each key from its value with `:`. And we separate pairs with `,`
- Every key must have a value. Every value must have a key. There is no such thing as a key with no value at all, and certainly there's no such thing as a value without a key.
- The keys must be unique within a dict.
- The keys must be *immutable*. That means you can use numbers and strings (and even tuples) as dict keys. But you cannot use lists or dicts.
- The values have no restrictions on them whatsoever. They can repeat, they can be mutable, they can be of any types you want, etc.
