# Dicts

A dictionary is Python's term for something that's very famous and well known in the programming world. It usually has other names, though:

- Hash table
- Hash map
- Map
- Name-value pair
- Key-value pair
- Associative array

The idea in a dict is that you aren't storing a single value. Rather, you're storing a *key* (the index, basically) and the value.

# Rules for a dict

- You create a dict with `{}`. This is the empty dict.
- If you want, you can create the dict with one or more key-value pairs.
    - `{'a':10}`
    - `{'a':10, 'b':20}`
- As you can see, we separate each key from its value with `:`. And we separate pairs with `,`
- Every key must have a value. Every value must have a key. There is no such thing as a key with no value at all, and certainly there's no such thing as a value without a key.
- The keys must be unique within a dict.
- The keys must be *immutable*. That means you can use numbers and strings (and even tuples) as dict keys. But you cannot use lists or dicts.
- The values have no restrictions on them whatsoever. They can repeat, they can be mutable, they can be of any types you want, etc.

# Exercise: Restaurant

We're going to define a dict containing a (very small, limited) restaurant menu. The items on the menu will be the keys (as strings) and the values will be integers (the prices). You're going to ask the user repeatedly what they want to order:

- If they enter an empty string, then stop asking and print the `total` of their bill.
- If they ask for something on the menu (i.e., a key in the `menu` dict), then add the price to the total, and print the price, the new total, and the item
- If they ask for something that is *not* on the menu (i.e., not a key in the `menu` dict), then scold them and let them try again

Example:

    Order: sandwich
    sandwich is 10, total is 10
    Order: tea
    tea is 7, total is 17
    Order: elephant
    sorry, we are fresh out of elephant today!
    Order: [ENTER]
    your total is 17

You'll want to start the program by defining a dict called `menu`. Then you can use it in the rest of the program.
    

```python
# you can think of the "menu" dictionary I'm going to create here as being a read-only
# database. I'll define it at the top of the program, and I'll use it.

menu = {'sandwich':10,  'tea':7,   'apple':3,   'cake':8}
total = 0

while True:
    order = input('Order: ').strip()
    
    if order == '':    # did we get the empty string?
        break

    if order in menu:        # if the user's input is a key in our dict
        price = menu[order]  # get the price
        total += price       # add price to total
        print(f'{order} costs {price}, total is now {total}')
    else:
        print(f'We are out of {order} today!')

print(f'total = {total}')
```

# Does the order of definition matter?

YES! When you define a dict, each pair is in the order of `KEY:VALUE`

If you have multiple key-value pairs, then they'll be `{KEY1:VALUE1, KEY2:VALUE2, etc.}`

This is important because:
- You retrieve values via the keys
- You cannot retrieve keys via the values
- It's a one-way street, and everything in a dict is done via the keys

```python
while True:   # the while loop's block will repeat FOREVER, ending only with "break"

    name = input('Enter your name: ').strip()

    if name == '':    # this ensures that the while-True loop has a chance to exit
        break

    print(f'Hello, {name}!')   
```

# Are dicts mutable?

Yes! You might remember, when we spoke about lists, we said that there are three ways that mutable data can be changed:

1. We can modify an existing value
2. We can add a new value
3. We can remove an existing value

```python
# how do I modify an existing value?
# just assign to it

d = {'a':10, 'b':20, 'c':30}

d['b'] = 999 # this replaces the existing value for 'b' with 999, an integer
```

```python
d
```

```python
# how can I add a new key-value pair to the dict?
# with a list, we need to use list.append or +=
# with a dict, we ... assign to the key we want!

d['x'] = 888   # because the key 'x' doesn't (yet) exist, it is added with value 888
d
```

```python
# how about removing a key-value pair?
# we do that with the "pop" method, specifying a key
# the key-value pair is removed, and the value for that key is returned.

d
```

```python
d.pop('b')   # this will remove the 'b':999 key-value pair, and return 999
```

# Exercise: Vowels, digits, and others (dict edition)

1. Define a dict, `counts`, with three key-value pairs. The keys will be `vowels`, `digits`, and `others`. The values will all be 0.
2. Ask the user to enter a text string.
3. Go through the string, one character at a time:
    - If it's a vowel (a,e,i,o,u), then add 1 to `counts['vowels']`
    - If it's a digit (0-9), then add 1 to `counts['digits']`
    - Otherwise, add 1 to `counts['others']`
4. In the end, print the `counts` dict.

```python
# this program uses what I call "paradigm #2" for dict use, meaning:
# we set up the keys and values, with all values set to an initial starting point.
# We don't add or remove keys during the program, but we do update the values.

counts = {'vowels':0,
          'digits':0,
          'others':0}

text = input('Enter text: ').strip()

for one_character in text:          # go through the user's input text one char at a time
    if one_character in 'aeiou':    # check: is this character a vowel?
        counts['vowels'] += 1       #   if so, add 1 to the value for counts['vowels']

    elif one_character.isdigit():   # check: is this character a digit?
        counts['digits'] += 1       #   if so, add 1 to the value for counts['digits']

    else:    # in all other cases...
        counts['others'] += 1

print(counts)        
```

# Exercise: Vowels, digits, and others (dict-list edition)

1. Repeat the previous exercise, except that instead of initializing all the values to be `0`, to count how many vowels, digits, and/or others you have, set each to be an empty list.
2. As you encounter a vowel, digit, or other, use the `list.append` method to add that character to the end of the appropriate list.
3. Over the course of the program, the keys will remain the same. The values (the lists) will grow, as characters of each type are found.
4. In the end, again print `counts`, the dict.
5. 

```python
counts = {'vowels':[],
          'digits':[],
          'others':[]}

text = input('Enter text: ').strip()

for one_character in text:          # go through the user's input text one char at a time
    if one_character in 'aeiou':    # check: is this character a vowel?
        counts['vowels'].append(one_character)       #   if so, add 1 to the value for counts['vowels']

    elif one_character.isdigit():   # check: is this character a digit?
        counts['digits'].append(one_character)       #   if so, add 1 to the value for counts['digits']

    else:    # in all other cases...
        counts['others'].append(one_character)

print(counts)        
```

# The story so far

1. We can define a dict and keep it as is in a program, treating it as a read-only database
2. We can define a dict with keys and initial values, and then modify/update those values

Paradigm 3 means: We start with nothing, with an empty dict

- If we see a key that we already have, we update the value (like paradigm 2)
- If we see a new key, we add it to the dictionary, with an initial value

This is good in situations where we don't know what keys we'll get, and we don't know what values we'll get... but we do know what to do with each when we get them.

```python
# character counter -- count how many times each character appears in a string

counts = {}   

text = input('Enter text: ').strip()

for one_character in text:

    # have we seen this character before?
    # we can check that with "in" on the dict -- if it's a key, we'll get True
    if one_character in counts:
        counts[one_character] += 1   # add 1 to the value for one_character

    else:
        counts[one_character] = 1    # first time seeing one_character? set the value to 1

print(counts)        
```
