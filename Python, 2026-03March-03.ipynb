# Intro to Jupyter

Tradtionally, we write Python code in a file, and then we run/execute the contents of that file. That's great, but not when we just want to try thing out.

For that, we use a REPL -- a read, eval, print loop. It allows you to try things out, and experiment, and see what your code does. 

Jupyter is a Web-based REPL. It pretends that Python is running inside of your browser. It gives you such an effective illusion, that it's a HUGELY popular project, especially with data scientists. 

You can have, in one notebook:

- Python code
- Markdown documentation (which is what I'm writing right now)
- Visualization (plo)
