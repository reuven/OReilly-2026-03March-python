# Variables and assignment

It's boring to have the same values all of the time! We have to repeat them, and that's considered bad. 

I can assign a value to a name, and then use that name repeatedly. In programming, we call those names *variables*.

We give a variable a value with the assignment operator, which is `=`.

**THE `=` in PYTHON HAS NOTHING TO DO WITH `=` in MATHEMATICS.**

In math, you would say

    2 + 5 = 7

That is *NOT* valid Python! 

When you assign in Python, you have a variable/name on the left, and a value on the right. The `=` between them tells Python: Take the value on the right, and assign to the variable on the left.

```python
x = 10   # x refers to the integer 10
y = 20   # y refers to the integer 20
```

```python
print(x + y)  # Python checks -- what is x? 10. what is y? 20. It adds them, and gives 30 to print
```

# What about declaring the variable?

```python
name = 'Reuven'   # I'm assigning a text value, 'Reuven', to the variable name
print('Hello, ' + name + '!')
```

# Conditionals

So far, everything we've written has gotten executed. If we have 10 lines of code, then all 10 lines will run.

But we don't always want that. In fact, a big part of writing a program is deciding which part will be run under which circumstances. 

This is known as "conditional execution," and it's fundamental to programming! We handle this with the keywords `if` and `else`.



# What's in this code?

1. We ask the user to enter their name. We assign that value to the variable *name*.
2. The `if` looks to its right, and checks if there is a `True` value there. If so, then its block executes.
3. At the end of the `if` line, there is a mandatory `:` (colon). That means: Starting on the next line, we'll have a block of code.
4. In Python, a block is always indented! That is mandatory! There are no `{}` like you might have in some other programming languages. In Python, any indentation is OK, but you must be consistent. For this reason, everyone (!) uses 4 spaces. For this reason, Jupyter (and other editors) will automatically indent your code in a block.
5. A block can contain any number of lines. They all must be indented. To stop the indentation and return to the outer layer, just press backspace.
6. On line 6, we have `else:`, whose block runs if the `if` block did not.
7. One of the blocks is guaranteed to run. Not zero, and not two.

```python
name = input('Enter your name: ')

if name == 'Reuven':
    print('Hi, boss!')
    print('Great to see you again today')
else:
    print('Hello, ' + name + '. Who are you?')
```

# More complex stuff

What if we have more than two options (for `if` and `else`?)

Then we can use `elif`. This comes after `if`, but before `else`. It also has a condition. The first condition that returns `True` has its block run. If none of the conditions are `True`, then `else` runs.

(You don't have to have an `else` block, but I think it's a really good idea most of the time.)

```python
name = input('Enter your name: ')

if name == 'Reuven':
    print('Hi, boss!')
    print('Great to see you again today')

elif name == 'someone else':  # as many elif blocks as I want/need!
    print('That is a very odd name, no?')

else:
    print('Hello, ' + name + '. Who are you?')
```
