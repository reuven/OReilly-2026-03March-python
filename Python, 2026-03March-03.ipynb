# Intro to Jupyter

Tradtionally, we write Python code in a file, and then we run/execute the contents of that file. That's great, but not when we just want to try thing out.

For that, we use a REPL -- a read, eval, print loop. It allows you to try things out, and experiment, and see what your code does. 

Jupyter is a Web-based REPL. It pretends that Python is running inside of your browser. It gives you such an effective illusion, that it's a HUGELY popular project, especially with data scientists. 

You can have, in one notebook:

- Python code
- Markdown documentation (which is what I'm writing right now)
- Visualization (plots + charts)
- Data analysis


# How to use Jupyter

Everything in Jupyter is done in "cells," like the one I'm typing into right now. When you type into a cell, you're in one of two modes:

- *EDIT* mode, where whatever you type goes into the cell. To get into edit mode, click inside of the cell or press ENTER.
- *COMMAND* mode, which means that what you type goes to Jupyter, and affects what it does. To get into command mode, either click to the left of the cell or press ESC. (The cell will then look a bit different.)

What commands can we use in this command mode? They're all one-character commands:

- `c` -- copy the current cell
- `x` -- cut the current cell
- `v` -- paste the most recently copied or cut cell
- `a` -- add a new, empty cell **ABOVE** the current one
- `b` -- add a new, empty cell **BELOW** the current one
- `y` -- make the current cell sensitive to Python code and able to run it
- `m` -- make the current cell for Markdown documentation

# What is a programming language? What is Python?

When computers were invented, each computer could solve one problem. If you wanted to solve another problem, you built a separate computer.

It was a huge step forward when people created *programmable* computers, that could solve many different problems. You would program them with 1s and 0s ("binary code"). That was awful and tedious! 

Pretty soon, people created *programming languages*, things that looked closer to English, but were then translated into 1s and 0s, and could run on the computer.

There are many *many* programming languages out there:
- C -- very close to this binary code, very fast to execute, but also very hard to write
- Java -- closer to people, but still complex
- Python -- a very high-level language, so meant for it to be easy for people

Python is not designed to be super speedy. It is designed to be easy to learn and easy to maintain programs that have already been written, because they are fairly readable.

Python is the perfect language for an age when people are expensive, and computers are cheap. Your time is important! Your time writing, and your time maintaining.

Python has been around for more than 30 years. It's dominant in a number of areas:
- Data science and machine learning
- Data analysis
- Devops (server configuration and distirbution)
- Automated testing
- Web applications
- Education

# What was this?

- `print` is a function. That's a verb in Python that does something. The `print` function displays something on the screen.
- To execute the function in Python, you must use `()`. They go right after the function's name.
- Inside of the `()`, you put one or more *arguments*, values that we want `print` to display. In this case, we only asked to display one thing. Every function takes some number of arguments.
- Our argument is a text value, known as a "string." Strings in Python need to have quotes around them, either single quotes or double quotes. If you don't put quotes around a string, Python thinks that it has to look up a name (function or a variable), and you'll get an error.

```python
print('a')
print('b')
print('c')
print('d')
```

```python
# I can print numbers, too! Not just text

print(20)
print(10)
```

```python
# I can also print the result of an expression -- meaning, something that gives me a value back

print(10 + 20)    # first, Python runs 10+20, gets 30, and print then prints that
```
