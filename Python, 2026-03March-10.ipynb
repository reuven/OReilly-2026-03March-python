# Methods

Methods, like functions, are verbs in Python. When you invoke a function, or when you invoke a method, you're asking for Python to *do* something. What it does, of course, depends on the definition of the function/method.

You can think of methods as functions that have been assigned to, or linked to, one particular data structure. Methods are defined inside of the definition of a data structure.

`len` is a function. It doesn't belong to any one particular data type. As a result, you say

    len('abcd')

or

    s = 'abcd'
    len(s)

In the case of a method, though, it *belongs* to a data type, such as `str`. As a result, the `str.upper` method in Python can only be invoked with a string. So you can say:

    'abcd'.upper()   # notice, we still need the () to run the method

or

    s = 'abcd'
    s.upper()

There are many more methods than functions in Python, because it helps to organize our code. (Or more realistically, the code of the people who have defined the data type.)    

What are the methods you can run on a particular type?

1. Go to the Python documentation, and read up on that type.
2. In Jupyter, start a method's name and then press TAB. That will give you a list of completions. It won't tell you what the methods do, but it'll at least give you their names.

How can you tell the difference between a function a method? Methods belong to data structures, and thus always come after a `.`. Functions typically don't, unless they're stored in a module (which we'll talk about in week 5).

# Why `if`/`elif`/`else` vs. `if`/`if`/`if`?

The difference is logical -- when you have an `if`/`elif`/`else` set of clauses, one and only one of those can fire! Precisely one will fire -- the first for which the condition is `True`.

However, if you have multiple `if` statements in a row, each is evaluated independently! Zero might fire, and all might fire!

```python
x = 10

if x > 5:
    print('Greater than 5')
if x > 7:
    print('Greater than 7')
if x > 9:
    print('Greater than 9')
if x > 11:
    print('Greater than 11')
```

```python
x = 10

if x > 5:
    print('Greater than 5')
elif x > 7:
    print('Greater than 7')
elif x > 9:
    print('Greater than 9')
elif x > 11:
    print('Greater than 11')
    
```

# Loops

Let's say that I have a string, and I want to print every character in the string:

```python
s = 'abcd'

# we can print one character with [i], where i is an index in the string

print(s[0])
print(s[1])
print(s[2])
print(s[3])
```

Unfortunately, this works!

It's repetitive code, which means that it takes longer to write, longer to read, longer to understand, longer to execute, and it also means that when (not if!) we have to maintain/change our code, we'll have to change more than one line, even though there's a lot of similarity there.

# DRY -- don't repeat yourself!

This is a famous statement among programmers, and it basically comes down to the idea that if you can reduce repetition, you'll win, both now and in the future.

Instead of repeating our code, we can use a *loop*. A loop allows us to have the computer repeat some instructions, typically with some small changes each time it handles an instruction.

There are two kinds of loops in Python:

- `for` loops, which repeat the same action once for each element in a sequence,
- `while` loops, which repeat the same action so long as a condition is `True`

`for` loops are far more common in Python, but we'll talk about `while` loops, too.

```python
# let's re-implement what we had before, using a for loop
# Python's loops tend to be fairly straightforward, at least at first

s = 'abcd'

for one_character in s:
    print(one_character)
```

# `for` loop syntax

1. Start with the keyword `for`.
2. Next comes the "loop variable," whose values will be changed with each iteration of the loop. In this case, our loop variable is `one_character`. The name of this variable is *completely* up to you, and has no effect on what is assigned in each iteration of the loop.
3. Then we have the word `in`.
4. Finally, on the first line, we have the value over which we'll be iterating -- in this case, a string, `s`.
5. At the end of the line, we have a `:`, because then comes an indented block.
6. In the loop body, you can have *any* code you want! `print`, `input`, another `for` loop, `if` with conditions... you name it!

# How a Python `for` loop works

1. `for` turns to `s`, the value at the end of the line, and asks: Are you iterable? Meaning, do you know how to behave inside of a `for` loop?
    - If not, then the loop exits with error.
2. Given that the value `s` is iterable, `for` asks it for its next value.
    - If there are no more values to give, `s` says so, and the `for` loop exits (silently).
3. The next value is assigned to the loop variable, `one_character`.
4. The loop body is executed, from start to finish.
5. When we get to the end of the loop body, Python returns to step 2, asking for the next value.

This means that the `for` loop has no way of knowing (a) how many iterations it'll run or (b) what types of values it'll receive.

# Notice!

- There's no index
- We don't know how long we'll be running
- We don't ask the value (`s`) for anything specific, such as from a particular index. We just say, "Give us your next thing."

The smarts of the loop are located inside of the value, not in the `for` loop.

All strings are iterable, as are lists, tuples, dicts, and files (among other things).

But some values are *not* iterable, such as integers.

```python
for one_value in 5:
    print(one_value)
```

In the case of a string, it's smart enough to know how to behave inside of a `for` loop. It gives us one character at a time.

My use of `one_character` as the loop variable name was not meant to influence what we'll get. (It has no such influence!) However, it was meant to make the code more readable by people (like me) who can always use appropriate variable names.

Changing the loop variable to `one_digit` or `one_terabyte` or `teacup`, will have zero influence on what we get with each iteration.

```python
# can we iterate over a 1-character string?
s = 'a'

print('Start')
for one_character in s:
    print(one_character)
print('End')    
```

```python
# can we iterate over a 0-character string?
s = ''   # empty string!

print('Start')
for one_character in s:
    print(one_character)
print('End')    
```
